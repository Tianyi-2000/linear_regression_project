```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from pygam import LinearGAM, s, f  # s = spline term, f = factor (categorical)
```

```python
sat = pd.read_csv('data/satgpa.csv')
sat.head()
```

```python
X = sat['sat_sum'].values.reshape(-1,1)
y = sat['fy_gpa']

model = LinearRegression()
model.fit(X, y)

x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_pred = model.predict(x_range)

plt.figure(figsize=(6,4))
plt.scatter(X, y, alpha=0.7)
plt.plot(x_range, y_pred, color="black", linewidth=2)

plt.title("High School GPA vs First Year GPA")
plt.xlabel("High School GPA")
plt.ylabel("First Year GPA")
plt.legend()
plt.grid(True)

plt.savefig("regression_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```
